```
import pandas as pd
import numpy as np

# Load (assuming df already loaded)
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str))
df.drop(columns=['Date', 'Time', 'Vehicle Images'], inplace=True)
df['Booking_ID'] = df['Booking_ID'].str.strip()  # Clean IDs
df.drop_duplicates(subset=['Booking_ID'], inplace=True)

# Success rides have TAT/ratings; fill 0 for others
success_cols = ['V_TAT', 'C_TAT', 'Payment_Method', 'Driver_Ratings', 'Customer_Rating']
df[success_cols] = df[success_cols].fillna(0)

# Fill reasons as empty string
df['Canceled_Rides_by_Customer'].fillna('', inplace=True)
df['Canceled_Rides_by_Driver'].fillna('', inplace=True)
df['Incomplete_Rides'].fillna('No', inplace=True)
df['Incomplete_Rides_Reason'].fillna('', inplace=True)

# Feature engineering for insights
df['Is_Success'] = (df['Booking_Status'] == 'Success').astype(int)
df['Total_Rating'] = (df['Driver_Ratings'] + df['Customer_Rating']) / 2
df['Revenue'] = df['Booking_Value'] * df['Is_Success']  # Only successful contribute revenue

print('Cleaned shape:', df.shape)
print(df['Booking_Status'].value_counts())
print(df.isnull().sum())
print(df.head())

df.to_csv('ola_cleaned.csv', index=False)
print('Exported to ola_cleaned.csv')
```

```
import sqlite3
import pandas as pd

# Load cleaned CSV to SQLite
df = pd.read_csv('/content/ola_cleaned.csv')
conn = sqlite3.connect('ola_rides.db')
df.to_sql('rides', conn, if_exists='replace', index=False)

# Verify
print('Rows loaded:', pd.read_sql('SELECT COUNT(*) FROM rides;', conn).iloc[0,0])
print('Status counts:\n', pd.read_sql("SELECT Booking_Status, COUNT(*) FROM rides GROUP BY Booking_Status;", conn))
print('Vehicle types:\n', pd.read_sql("SELECT Vehicle_Type, COUNT(*) FROM rides GROUP BY Vehicle_Type ORDER BY COUNT(*) DESC;", conn))

# Query 1 sample
q1 = "SELECT * FROM rides WHERE Booking_Status = 'Success' LIMIT 5;"
print('\nQuery 1:\n', pd.read_sql(q1, conn))
conn.close()
print('SQLite ready! Download ola_rides.db for Power BI.')
```

Query 1: All Successful Bookings


```
import sqlite3
import pandas as pd
conn = sqlite3.connect('ola_rides.db')

q1 = """
SELECT COUNT(*) as Success_Count, AVG(Revenue) as Avg_Revenue
FROM rides
WHERE Booking_Status = 'Success'
"""
df1 = pd.read_sql(q1, conn)
print(df1)
df1.to_csv('Q1_Successful.csv', index=False)
conn.close()
```

Query 2: Average Ride Distance per Vehicle Type

```
conn = sqlite3.connect('ola_rides.db')
q2 = """
SELECT Vehicle_Type,
       ROUND(AVG(Ride_Distance), 2) as Avg_Distance_km
FROM rides
GROUP BY Vehicle_Type
ORDER BY Avg_Distance_km DESC
"""
df2 = pd.read_sql(q2, conn)
print(df2)
df2.to_csv('Q2_AvgDist_Vehicle.csv', index=False)
```

Query 3: Total Customer-Cancelled Rides

```
conn = sqlite3.connect('ola_rides.db')
q3 = """
SELECT COUNT(*) as Customer_Cancelled_Count
FROM rides
WHERE Booking_Status = 'Canceled by Customer'
"""
df3 = pd.read_sql(q3, conn)
print(df3)
df3.to_csv('Q3_Cust_Cancels.csv', index=False)
```

Query 4: Top 5 Customers by Ride Count

```
conn = sqlite3.connect('ola_rides.db')
q4 = """
SELECT Customer_ID, COUNT(*) as Ride_Count
FROM rides
WHERE Is_Success = 1
GROUP BY Customer_ID
ORDER BY Ride_Count DESC
LIMIT 5
"""
df4 = pd.read_sql(q4, conn)
print(df4)
df4.to_csv('Q4_Top5_Customers.csv', index=False)
```

Query 5: Driver Cancellations (Personal/Car Issues)

```
conn = sqlite3.connect('ola_rides.db')
q5 = """
SELECT
  COUNT(*) as Total_Driver_Cancels,
  SUM(CASE WHEN Canceled_Rides_by_Driver LIKE '%Personal%' OR Canceled_Rides_by_Driver LIKE '%Car%' THEN 1 ELSE 0 END) as Personal_Car_Issues
FROM rides
WHERE Booking_Status IN ('Canceled by Driver', 'Driver Not Found')
"""
df5 = pd.read_sql(q5, conn)
print(df5)
df5.to_csv('Q5_Driver_Cancels.csv', index=False)
```

Query 6: Max/Min Driver Ratings for Prime Sedan

```
conn = sqlite3.connect('ola_rides.db')
q6 = """
SELECT
  MAX(Driver_Ratings) as Max_Driver_Rating,
  MIN(Driver_Ratings) as Min_Driver_Rating
FROM rides
WHERE Vehicle_Type = 'Prime Sedan' AND Driver_Ratings > 0
"""
df6 = pd.read_sql(q6, conn)
print(df6)
df6.to_csv('Q6_PrimeSedan_Ratings.csv', index=False)
```

Query 7: All UPI Payment Rides (Count + Sample)

```
conn = sqlite3.connect('ola_rides.db')
q7 = """
SELECT COUNT(*) as UPI_Rides_Count
FROM rides
WHERE Payment_Method = 'UPI' AND Is_Success = 1
"""
df7 = pd.read_sql(q7, conn)
print(df7)
df7.to_csv('Q7_UPI_Rides.csv', index=False)
```

Query 8: Average Customer Rating per Vehicle Type

```
conn = sqlite3.connect('ola_rides.db')
q8 = """
SELECT Vehicle_Type,
       ROUND(AVG(Customer_Rating), 2) as Avg_Cust_Rating
FROM rides
WHERE Is_Success = 1 AND Customer_Rating > 0
GROUP BY Vehicle_Type
ORDER BY Avg_Cust_Rating DESC
"""
df8 = pd.read_sql(q8, conn)
print(df8)
df8.to_csv('Q8_AvgCustRating_Vehicle.csv', index=False)
```

Query 9: Total Booking Value (Successful Rides)

```
conn = sqlite3.connect('ola_rides.db')
q9 = """
SELECT SUM(Revenue) as Total_Revenue
FROM rides
WHERE Is_Success = 1
"""
df9 = pd.read_sql(q9, conn)
print(df9)
df9.to_csv('Q9_Total_BookingValue.csv', index=False)
```

Query 10: All Incomplete Rides + Reasons

```
conn = sqlite3.connect('ola_rides.db')
q10 = """
SELECT Booking_Status,
       Incomplete_Rides_Reason,
       COUNT(*) as Count
FROM rides
WHERE Booking_Status != 'Success'
GROUP BY Booking_Status, Incomplete_Rides_Reason
ORDER BY Count DESC
"""
df10 = pd.read_sql(q10, conn)
print(df10)
df10.to_csv('Q10_Incomplete_Rides.csv', index=False)
conn.close()
```
